```

import random
import datetime
time=datetime.datetime.now().hour
def greetings():
    l=[]
    l.append("Hello...! , May i Know Your Name..")
    l.append("Hi Buddy..! This is bot.. What's your Name..")
    print(random.choice(l)+"ðŸ¤—")

def INTRODUCE(name):
    #name=input("Enter your name :")
    greet="Good Morning..!"
    if time>20:
        greet="Good Evening..!"
    elif time>15:
        greet="Good Evening..!"
    elif time>=11:
        greet="Good Afternoon..!"
    print(greet+" "+name+", I can u help to book ur movie ticket..ðŸŽ¥")

def DATE():
    print("-----------------------")
    print("Choose the day...!")
    print("-----------------------")
    print("1. TODAY , AFTER",time,": 00")
    print("2. Tomorrow.")
    print("3. Day After Tomorrow.")
    n=int(input("Enter the number from 1,2 and 3:\n"))
    if n not in [1,2,3]:
        print("-----------------------")
        print("Enter Valid Input..!ðŸ˜¥")
        print("-----------------------")
        DATE()
    return n

def THEATRE():
    print("-----------------------")
    print("SELECT A THEATRE...!")
    print("-----------------------")
    for i in range(1,6):
        print(i,"."," THEATRE ",i,sep="")

def MOVIE():
    print("-----------------------")
    print("SELECT A MOVIE...!")
    print("-----------------------")
    print("1. AVENGERS ENDGAME")
    print("2. THE DARK KNIGHT RISESðŸ–¤")
    print("3. SPIDERMAN - FAR FROM HOMEðŸ¤˜")

def MOVIE_NAME(t):
    if t==1:
        return "AVENGERS ENDGAME"
    elif t==2:
        return "THE DARK KNIGHT RISESðŸ–¤"
    else:
        return "SPIDERMAN - FAR FROM HOMEðŸ¤˜"

def SHOWS(n):
    l=[]
    print("-----------------------")
    print("SELECT SHOW TIMINGS...!")
    print("-----------------------")
    l.append("1. 11 A.M.")
    l.append("2. 2:00 P.M")
    l.append("3. 5:30 P.M.")
    l.append("4. 9:30 P.M.")
    if n==1:
        if time<11:
            for i in l:
                print(i) 
        elif time<14:
           for i in l[1:]:
                print(i) 
                return 1
        elif time<17:
           for i in l[2:]:
                print(i)  
                return 2
        elif time<21:
           for i in l[3:]:
                print(i)   
                return 3  
        else:
            print("SORRY...! NO SHOW AVAILABLE TODAY CHOOSE OTHER DAY..!") 
            SHOWS(n)
    else:      
        for i in l:
            print(i)
        return 1

def SHOW_TIMINGS(i):
    if i==1:
        return "11:00 A.M"
    elif i==2:
        return "2:00 P.M"
    elif i==3:
        return "5:30 P.M."
    else:
        return "9:30 P.M."

def TICKET(t,NO_OF_TICKETS,s,show,m):
    print("YOUR ORDER IS :")
    print("*****************************")
    print("MOVIE NAME        : ",MOVIE_NAME(m))
    print("THEATRE_NAME      :  THEATRE",t)
    print("NUMBER OF TICKETS : ",NO_OF_TICKETS)
    print("ON ",s,SHOW_TIMINGS(show))
    print("*****************************")
    print("DO YOU WANT TO MAKE ANOTHER ORDER :")
    print("1. YES")
    print("2. NO")
    if (input("ENTER YOUR OPINION :")=="1"):
        bot()
    else:
        print("THANK YOU VERY MUCH FOR YOUR ORDER , VISIT AGAIN...!â˜º")

def bot():   
    greetings()
    name=input("Enter your name :")
    INTRODUCE(name)
    n=DATE()
    if n==1:
        s="Today "
    elif n==2:
        s="Tomorrow"
    else:
        s="Day After Tomorrow"
    MOVIE()
    m=int(input("ENTER MOVIE OPTION WITH NUMBERS 1,2 & 3 :"))
    if m not in [1,2,3]:
        print("-----------------------")
        print("Enter Valid Input..!ðŸ˜¥")
        print("-----------------------")
        MOVIE()
        m=int(input("ENTER MOVIE OPTION WITH NUMBERS 1,2 & 3 :"))
    THEATRE()
    t=int(input("ENTER THEATRE OPTION :"))
    while t not in range(1,6):
        print("-----------------------")
        print("Enter Valid Input..!ðŸ˜¥")
        print("-----------------------")
        THEATRE()
        t=int(input("ENTER THEATRE OPTION :"))
    q=SHOWS(n)
    show=int(input("ENTER SHOW TIMINGS OPTION :"))
    while show not in range(q,5):
        print("-----------------------")
        print("Enter Valid Input..!ðŸ˜¥")
        print("-----------------------")
        SHOWS(n)
        show=int(input("ENTER SHOW TIMINGS OPTION :"))
    NO_OF_TICKETS = int(input("ENTER NUMBER OF TICKETS..!"))
    TICKET(t,NO_OF_TICKETS,s,show,m)

bot()
```
